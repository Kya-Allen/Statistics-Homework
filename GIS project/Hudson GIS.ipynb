```python
NearestNeighbor = []
for i in range(0,770):
    Distances = []
    for j in range (0,769):
        if i != j:
            Distances.append(np.linalg.norm(Grabs['geometry'][i] - Grabs['geometry'][j]))
    NearestNeighbor.append(min(Distances))    
NearestNeighbor        
```

```python
Grabs['NN'] = NearestNeighbor
```

```python
Grabs['NN'].mean()
```
