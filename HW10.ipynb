```python
import numpy as np
import pandas as pd
```

```python
Averaged_df = pd.read_csv('AveragedData.csv')
Averaged_df = Averaged_df.set_index('Monitoring station')
Averaged_df
```

## The above dataframe is from a CSV of the 3 year averages for each data point, computed in Google Sheets. Now We'll Normalize the vectors with L1 (Manhattan) and L2 (Euclidean) norms to make sure the scale of all the data is comparable for distance based metrics.

### L2 (Euclidean) Norm - values divided by the square root of the sum of the squared values

```python

AveragedMatrix = np.array(Averaged_df)
AveragedMatrix = AveragedMatrix.T

for i in range(0, 5):
    AveragedMatrix[i] = np.sqrt(np.sum(np.square(AveragedMatrix[i])))
AveragedMatrix
```

```python
L2NormMatrix = np.array(Averaged_df.T)
L2NormMatrix = np.divide(L2NormMatrix, AveragedMatrix)
L2Norm_df = pd.DataFrame(L2NormMatrix.T)
L2Norm_df
```

### L1 (Manhattan) Norm - Values divided by the sum of the absolute values

```python
AveragedMatrix = np.array(Averaged_df)
AveragedMatrix = AveragedMatrix.T

for i in range(0, 5):
    AveragedMatrix[i] = np.sum(np.absolute(AveragedMatrix[i]))
AveragedMatrix
```

```python
L1NormMatrix = np.array(Averaged_df.T)
L1NormMatrix = np.divide(L1NormMatrix, AveragedMatrix)
L1Norm_df = pd.DataFrame(L1NormMatrix.T)
L1Norm_df
```

### Now, Calculate weighted decision matrices for L1 and L2, each with two as instructed (w = 1, w = proportion)
### Spoiler: Part B Weighting is already done. Weights of 1 don't change anything. Onto the next weighting scheme

```python
L1WeightMatrix = np.array(L1Norm_df.T)
L2WeightMatrix = np.array(L2Norm_df.T)

L1_Weights = []
L2_Weights = []

for i in range(0,5):
    L1_Weights.append(np.divide(np.sum(L1WeightMatrix[i]), (np.sum(L1WeightMatrix[0]) + np.sum(L1WeightMatrix[1])
                                                                    + np.sum(L1WeightMatrix[2]) + np.sum(L1WeightMatrix[3]) +
                                                                   np.sum(L1WeightMatrix[4]))))
L1_Weights
```

```python
for i in range(0,5):
    L2_Weights.append(np.divide(np.sum(L2WeightMatrix[i]), (np.sum(L2WeightMatrix[0]) + np.sum(L2WeightMatrix[1])
                                                                    + np.sum(L2WeightMatrix[2]) + np.sum(L2WeightMatrix[3]) +
                                                                   np.sum(L2WeightMatrix[4]))))
L2_Weights
```

```python
for i in range (0,5):
    L1WeightMatrix[i] = L1WeightMatrix[i] * L1_Weights[i]
    L2WeightMatrix[i] = L2WeightMatrix[i] * L2_Weights[i]
    
L1Weight_df = pd.DataFrame(L1WeightMatrix.T)
L2Weight_df = pd.DataFrame(L2WeightMatrix.T)
L1Weight_df
```

```python
L2Weight_df
```

## Now we compute ideal/negative-ideal row vectors, then compute the Euclidian and Manhattan distances from each real row, to them.

```python
#I'm assuming a greater value is always worse, since these are pollutants (so the Ideal row takes the Minimimums)
L2_IDR = []
L2_NIDR = []
L1_IDR = []
L1_NIDR = []

for i in range(0,5):
    L1_IDR.append(L1Weight_df[i].min())
    L1_NIDR.append(L1Weight_df[i].max())
    L2_IDR.append(L2Weight_df[i].min())
    L2_NIDR.append(L2Weight_df[i].max())
```

```python
L2_IDR
```

```python
L1_distancesIDR = []
L1_distancesNIDR = []
L2_distancesIDR = []
L2_distancesNIDR = []

for i in range(0,7):
    L1RowVector = np.array(L1Weight_df.loc[i])
    L1_distancesIDR.append(np.sum(np.absolute(L1RowVector - L1_IDR)))
    L1_distancesNIDR.append(np.sum(np.absolute(L1RowVector - L1_NIDR)))
    L2RowVector = np.array(L2Weight_df.loc[i])
    L2_distancesIDR.append(np.sqrt(np.sum(np.square(L2RowVector - L2_IDR))))
    L2_distancesNIDR.append(np.sqrt(np.sum(np.square(L2RowVector - L2_NIDR))))
```

```python
L1_Metric = np.array(np.divide(L1_distancesIDR, (np.array(L1_distancesIDR) + np.array(L1_distancesNIDR))))
L2_Metric = np.array(np.divide(L2_distancesIDR, (np.array(L2_distancesIDR) + np.array(L2_distancesNIDR))))
L1_Metric
```

```python
L2_Metric
```

## And again, with the unweighted versions

```python
L2_IDR_UW = []
L2_NIDR_UW = []
L1_IDR_UW = []
L1_NIDR_UW = []

for i in range(0,5):
    L1_IDR_UW.append(L1Norm_df[i].min())
    L1_NIDR_UW.append(L1Norm_df[i].max())
    L2_IDR_UW.append(L2Norm_df[i].min())
    L2_NIDR_UW.append(L2Norm_df[i].max())

L1_distancesIDR_UW = []
L1_distancesNIDR_UW = []
L2_distancesIDR_UW = []
L2_distancesNIDR_UW = []

for i in range(0,7):
    L1RowVector_UW = np.array(L1Norm_df.loc[i])
    L1_distancesIDR_UW.append(np.sum(np.absolute(L1RowVector_UW - L1_IDR_UW)))
    L1_distancesNIDR_UW.append(np.sum(np.absolute(L1RowVector_UW - L1_NIDR_UW)))
    L2RowVector_UW = np.array(L2Norm_df.loc[i])
    L2_distancesIDR_UW.append(np.sqrt(np.sum(np.square(L2RowVector_UW - L2_IDR_UW))))
    L2_distancesNIDR_UW.append(np.sqrt(np.sum(np.square(L2RowVector_UW - L2_NIDR_UW))))
    
L1_Metric_UW = np.array(np.divide(L1_distancesIDR_UW, (np.array(L1_distancesIDR_UW) + np.array(L1_distancesNIDR_UW))))
L2_Metric_UW = np.array(np.divide(L2_distancesIDR_UW, (np.array(L2_distancesIDR_UW) + np.array(L2_distancesNIDR_UW))))
```

## Now, doing the TOPSIS method on a "Rank Matrix", where simple ranks are pre-generated for each criteria. (computed in google sheets, and imported via CSV)

```python
RankMatrix = np.array(RankMatrix_df)
RankMatrix = RankMatrix.T

for i in range(0, 5):
    RankMatrix[i] = np.sqrt(np.sum(np.square((RankMatrix[i]))))
RankMatrix
```

```python
L2NormRank = np.array(RankMatrix_df.T)
L2NormRank = np.divide(L2NormRank, RankMatrix)
L2NormRank_df = pd.DataFrame(L2NormRank.T)
L2NormRank_df
```

## L1

```python
RankMatrix = np.array(RankMatrix_df)
RankMatrix = RankMatrix.T

for i in range(0, 5):
    RankMatrix[i] = np.sum(np.absolute(RankMatrix[i]))
RankMatrix

L1NormRankMatrix = np.array(RankMatrix_df.T)
L1NormRankMatrix = np.divide(L1NormRankMatrix, RankMatrix)
L1NormRank_df = pd.DataFrame(L1NormRankMatrix.T)
L1NormRank_df
```

## Rank Matrix TOPSIS-MCDM time

```python
L2_IDR_Rank = []
L2_NIDR_Rank = []
L1_IDR_Rank = []
L1_NIDR_Rank = []

for i in range(0,5):
    L1_IDR_Rank.append(L1NormRank_df[i].min())
    L1_NIDR_Rank.append(L1NormRank_df[i].max())
    L2_IDR_Rank.append(L2NormRank_df[i].min())
    L2_NIDR_Rank.append(L2NormRank_df[i].max())

L1_distancesIDR_Rank = []
L1_distancesNIDR_Rank = []
L2_distancesIDR_Rank = []
L2_distancesNIDR_Rank = []

for i in range(0,7):
    L1RowVector_Rank = np.array(L1NormRank_df.loc[i])
    L1_distancesIDR_Rank.append(np.sum(np.absolute(L1RowVector_Rank - L1_IDR_Rank)))
    L1_distancesNIDR_Rank.append(np.sum(np.absolute(L1RowVector_Rank - L1_NIDR_Rank)))
    L2RowVector_Rank = np.array(L2NormRank_df.loc[i])
    L2_distancesIDR_Rank.append(np.sqrt(np.sum(np.square(L2RowVector_Rank - L2_IDR_Rank))))
    L2_distancesNIDR_Rank.append(np.sqrt(np.sum(np.square(L2RowVector_Rank - L2_NIDR_Rank))))
    
L1_Metric_Rank = np.array(np.divide(L1_distancesIDR_Rank, (np.array(L1_distancesIDR_Rank) + np.array(L1_distancesNIDR_Rank))))
L2_Metric_Rank = np.array(np.divide(L2_distancesIDR_Rank, (np.array(L2_distancesIDR_Rank) + np.array(L2_distancesNIDR_Rank))))
```
